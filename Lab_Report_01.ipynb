```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
#task 1: Extract a 100×100 Region from the Center of the Image read from an URL.
# Load image from URL
# URL of the image
image_url = 'https://cdn.shopify.com/s/files/1/0274/9503/9079/files/20220211142754-margherita-9920_5a73220e-4a1a-4d33-b38f-26e98e3cd986.jpg?v=1723650067'

# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array

# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# Step 4: Get image dimensions and calculate center coordinates
height, width, _ = img.shape  # Get height, width, and channels of the image
y = height // 2  # Calculate the center y-coordinate
x = width // 2   # Calculate the center x-coordinate


# Step 4:Crop a 100x100 region from the center
crop = img[y-50:y+50, x-50:x+50]

# Step 5: Display the original image and the cropped image side by side using Matplotlib
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title("Cropped 100×100 Region")
plt.axis('off')

plt.show()
```

```python
# Task 2: Convert to HSV and show components

# Step 1: Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 2: Display the shape of the grayscale image
print(f"Shape of the Grayscale Image: {gray_image.shape}")

# Step 3: Visualize the grayscale image
plt.figure(figsize=(6, 6))
plt.title("Grayscale Image")
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()

# Step 4: Convert the image from BGR to HSV color space
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
hue, sat, val = cv2.split(hsv)

# Step 5: Create a figure for plotting the channels
plt.figure(figsize=(12, 4))
titles = ["Hue Channel", "Saturation Channel", "Value Channel"]
images = [hue, sat, val]


# Step 6: Loop through and display the channels side by side
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

# Step 7: Show the plot
plt.show()
```
